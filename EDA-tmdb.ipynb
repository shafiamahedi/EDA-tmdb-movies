## Performing Data cleaning and EDA on dataset

```python
#number of duplicate rows
sum(movies.duplicated())
```

```python
#delete duplicate rows
movies.drop_duplicates(inplace = True)
```

```python
print("After deleting duplicate rows : ",movies.shape)
```

```python
#drop columns that are not necessary
movies_df= movies.drop(['homepage','tagline', 'overview','budget_adj','revenue_adj'],axis=1)
```

```python
movies_df
```

```python
avg = int(np.mean(movies_df['budget']))
```

```python
movies_df['budget'] = movies_df['budget'].replace(0,avg)
```

```python
avg = int(np.mean(movies_df['revenue']))
```

```python
movies_df['revenue'] = movies_df['revenue'].replace(0,avg)
```

```python
movies_df
```

## 1. Which movie had the most profit during 2000-2005?

```python
profit = movies_df['revenue']-movies_df['budget']
movies_df['profit'] = profit
movies_df.head(5)
```

```python
p = movies_df[(movies_df.release_year>=2000)&(movies_df.release_year<=2005)]
p
```

```python
pro = p[p.profit==p.profit.max()]
print("Movie that had the most profit during 2000-2005 is\n")
pro.iloc[:,[5,16]]
```

## 2. Which are the movies with the second lowest and second highest budget?

```python
movies_df.sort_values('budget')
```

```python
df_bud = movies_df.drop_duplicates(subset='budget', keep="first")
```

```python
movies_df
```

```python
df_bud
```

```python
df_bud = df_bud.sort_values('budget')
df_bud
```

```python
print("The second lowest budget movie is\n")
df_bud.iloc[1:2,[3,5]]
```

```python
print("The second largest budget movie is\n")
df_bud.iloc[-2:-1,[3,5]]
```

## 3. Which are the movies with most and least earned revenue?

```python
print("Most earned revenue movie is\n")
max_rev = movies_df[movies_df.revenue == movies_df.revenue.max()]
max_rev.iloc[:,[4,5]]
```

```python
print("Least earned revenue movies are\n")
min_rev = movies_df[movies_df.revenue==movies_df.revenue.min()]
min_rev.iloc[:,[4,5]]
```

## 4. What is the average runtime of movies in the year 2000? 

```python
print("The average runtime of movies in the year 2000 is\n")
av = movies_df[movies_df.release_year==2000]
av["runtime"].mean()
```
